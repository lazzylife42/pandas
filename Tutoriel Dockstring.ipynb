## 04 Les filtres

```python
df["gender"] == "Male"
```

```python
df[df["gender"] == "Male"] # pour filtrer la df 
```

```python
female_filter = df["gender"] == "Female"  # plus propre
df[female_filter]
```

```python
country_filter = df["country"].isin(["France","Canada"]) 
df[country_filter]
```

```python
df_test =  df.copy()
df_test.price_paid = df_test.price_paid.apply(lambda x: x.replace("$", ""))
df_test.price_paid = df_test.price_paid.astype(float)
df_test[df_test["price_paid"] >= 5.0 ]
```
