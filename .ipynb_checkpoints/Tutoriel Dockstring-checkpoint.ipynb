## 05 Traiter les valeurs manquantes

```python
df = pd.read_csv("data.csv")
df.isnull() # pour voir si c'est NULL
```

```python
df.notnull()
```

```python
df[df["tax"].notnull()] # utiliser la func pour filtrer
```

```python
df["tax"].fillna(0) # pour fill les NULL
```

```python
df.dropna() # enlève tout ce qui est NULL
```

```python
df.dropna(subset=["tax"], inplace=True) # pour spécifier la cible
df
```

## 06 Ajouter des colonnes

```python
df = pd.read_csv("data.csv")
```

```python
df["tax"] = df["tax"].fillna(0)
df
```

```python
df.dtypes
```

```python
# ici on convertis pour faire des calcules avec le même type

df.price_paid = df.price_paid.apply(lambda x: x.replace("$", ""))
df.price_paid = df.price_paid.astype(float) 
```

```python
df
```

```python
df.dtypes
```

```python
# On calcule le prix payé total
df["price_total"] = df["price_paid"] * (1 - df["tax"] / 100)
```

```python
df
```

```python
# Ici on créer une colonne en mappant un dictionaire sur le bon champ

countries = {"United States" : "UN",
             "France" : "FR",
             "Canada" : "CA",
             "Moroco" : "MA"}

df["country_code"] = df["country"].map(countries)
df.head(10)
```
